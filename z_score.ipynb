```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Normal Distribution**

```
def pdf_normal_distribution(x):
  return np.exp(-1.0 * (x**2)/(2))/np.sqrt(2 * np.pi)
```

```
x = np.linspace(-4, 4, 1000)
plt.plot(x, pdf_normal_distribution(x))
plt.title('Normal Distribution')
plt.ylabel('Probability Density Function')
```

**Standard Normal Table**

```
from scipy.integrate import quad
standard_normal_table = pd.DataFrame(data = [],
                                     index = np.round(np.arange(0, 3.6, .1),2),
                                     columns = np.round(np.arange(0.00, .1, .01), 2))

for index in standard_normal_table.index:
    for column in standard_normal_table.columns:
        z = np.round(index + column, 2)
        value, _ = quad(pdf_normal_distribution, np.NINF, z)
        standard_normal_table.loc[index, column] = np.round(value,4)

standard_normal_table.index = standard_normal_table.index.astype(str)
standard_normal_table.columns = [str(column).ljust(4,'0') for column in standard_normal_table.columns]
```

**Negative Standard Normal Table**

```
from scipy.integrate import quad
n_standard_normal_table = pd.DataFrame(data = [],
                                     index = np.round(np.arange(-3.5, 0.1, .1),2),
                                     columns = np.round(np.arange(0.00, .1, .01), 2))

for index in n_standard_normal_table.index:
    for column in n_standard_normal_table.columns:
        z = np.round(index + column, 2)
        value, _ = quad(pdf_normal_distribution, np.NINF, z)
        n_standard_normal_table.loc[index, column] = np.round(value,4)

n_standard_normal_table.index = n_standard_normal_table.index.astype(str)
n_standard_normal_table.columns = [str(column).ljust(4,'0') for column in n_standard_normal_table.columns]
```

**Compute Z-score**

```
def z_score(x):
  return np.round(quad(pdf_normal_distribution, np.NINF, x)[0], 4)
```
